```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

1. Построим красивое сердечко, чтобы поднять настроение миру и проверяющему <3

```python
t = np.linspace(-20, 20, 1000)
x = 16 * np.power(np.sin(t), 3)
y = 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3 * t) - np.cos(4 * t)
```

```python
fig = plt.figure(figsize=(5, 5))
ax = plt.axes()

ax.set(xlim=(-20, 20), ylim=(-20, 20), xlabel='x', ylabel='f(x)', title='Нeart');
ax.plot(x, y, 'r')
ax.grid(color='grey', linestyle='--', linewidth=0.3)
```

4. Посмотрим как доступность интернета ('internet') влияет на успеваемость ('G3'). Можно увидеть, что интернет - творит небольшие чудеса с образованием студентов.

```python
fg = sns.FacetGrid(data, hue="internet", aspect=3, xlim=(0, 20))
fg.map(sns.kdeplot, "G3", shade=True).add_legend();
```

5. Посмотрим на распределение оценок 'G3' в зависимости от желания получать высшее образования (higher). Видно, что дети, получившие оценки выше имеют больше мотивации получать высшее образование.

```python
sns.boxplot(x="higher", y="G3", data=data);
```
