```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd

%config InlineBackend.figure_format = 'retina'
from pylab import rcParams
rcParams['figure.figsize'] = (12, 9)
```

```python
df = pd.read_excel('geo.xlsx')
df.drop('Unnamed: 0', axis = 1, inplace= True)
```

```python
df.head()
```

```python
#plt.scatter(df.x, df.y,c=df.comment_class, cmap='autumn', s=60)
plt.scatter(df.x, df.y, cmap='autumn', s=60)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[['x','y']])
```

```python
X_scaled
```

```python
#plt.scatter(X_scaled[:,0], X_scaled[:,1], c=df.comment_class, cmap='autumn', s=60)
plt.scatter(X_scaled[:,0], X_scaled[:,1], cmap='autumn', s=60)
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, AffinityPropagation
from sklearn.metrics import adjusted_rand_score, silhouette_score
```

Будем использовать k-means

```python
k_inertia = []
ks = range(1,11)

for k in ks:
    clf_kmeans = KMeans(n_clusters=k)
    clusters_kmeans = clf_kmeans.fit_predict(X_scaled, )
    k_inertia.append(clf_kmeans.inertia_)
```

```python
plt.plot(ks, k_inertia)
```

Экспертно оценим, что лучше всего использовать 4 кластера.

```python
algorithms = [
    #AffinityPropagation(),
    KMeans(n_clusters=4)
    #AgglomerativeClustering(n_clusters=2),
    #DBSCAN(),
]
```

```python
clusters_random = np.random.randint(low=0, high=4, size=len(X_scaled))
```

```python
def get_descr(algo_name, y, y_pred, X):
    return "{}\nARI {:.2f}\nSilhouette {:.2f}".format(
        algo_name,
        adjusted_rand_score(y, y_pred),
        silhouette_score(X, y_pred)
    )
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 2))
axes[0].scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters_random, cmap='autumn', s=60)
axes[0].set_title(get_descr("Random", df.comment_class, clusters_random, X_scaled))
for ax, algorithm in zip(axes[1:], algorithms):
    # кластеризуем и выводим картинку
    clusters = algorithm.fit_predict(X_scaled,)
    ax.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='autumn', s=60)
    ax.set_title(get_descr(algorithm.__class__.__name__, df.comment_class, clusters, X_scaled))
    
    # если есть центры кластеров - выведем их
    if algorithm.__class__.__name__ in {'KMeans', 'AffinityPropagation'}:
        centers = algorithm.cluster_centers_
        ax.scatter(centers[:, 0], centers[:, 1], s=50)
```
