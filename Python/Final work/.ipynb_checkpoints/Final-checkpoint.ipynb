3. Рассчитать и визуализировать корреляционную матрицу для переменных


```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(data.corr(), interpolation='nearest')
fig.colorbar(cax)
plt.show()
```

#### 9. Сколько сотрудников покинуло компанию. 

```python
data_group_salary=data.groupby(['left']).count()
```

```python
i = [0]
def absolute_value(val):
    a  = data_group_salary.iloc[i[0]%len(data_group_salary),i[0]//len(data_group_salary)]
    i[0] += 1
    return a

data_group_salary['salary'].plot(kind='pie', autopct=absolute_value , 
 startangle=90, shadow=False, legend = True, fontsize=14)
```
