```python
from sklearn.datasets import load_boston
import numpy as np
import pandas as pd
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
boston=load_boston()
data = load_boston()
df = pd.DataFrame(data.data, columns=data.feature_names)
df.head()
```

этап 2: взять два атрибута и посчитать скалярное произведение двух векторов (2 дома), начертить 2 дома как вектора в двухмерном пространстве, и сравниваем графически и через скалярное произведение.


```python
fig = plt.figure(figsize=(10, 10))


X = np.array(df['RM'])
Y = np.array(df['DIS'])
V = np.array([[X[0],Y[0]],[X[5],Y[5]]])
origin = [0], [0] # origin point
print(f'Скалярное произведение равно {np.dot(V[0],V[1])}')
plt.quiver(*origin, V[:,0], V[:,1], color=['r','b','g'], scale=21)
plt.show()
```

Нарисовать два дома в 3-мерном пространстве (атрибуты выбрать самостоятельно)

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.gca(projection='3d')

x = np.array(df['DIS'])
y = np.array(df['RAD'])
z = np.array(df['TAX'])

ax.plot([0,x[0]], [0,y[0]], [0,z[0]])
ax.plot([0,x[1]], [0,y[1]], [0,z[1]])
ax.plot([0,x[3]], [0,y[3]], [0,z[3]])

ax.scatter(x[0], y[0], z[0], c='b', marker='o')
ax.scatter(x[1], y[1], z[1], c='r', marker='o')
ax.scatter(x[3], y[3], z[3], c='g', marker='o')

ax.set_xlabel('weighted distances to five Boston employment centres')
ax.set_ylabel('index of accessibility to radial highways')
ax.set_zlabel('full-value property-tax rate per $10,000')

plt.show()
```

Найти косинус угла между векторами x1 и x2, где x1 и x2 - это два дома из 2ух атрибутов, полученных на этапе 2.

Найти расстояние между точками x1 и x2.

```python
from math import sqrt, acos, degrees
 
x1 = V[0][0]
y1 = V[0][1]

 
x2 = V[1][0]
y2 = V[1][1]

 
def scalar(x1, y1, x2, y2):
    return x1*x2 + y1*y2 
 
def module(x, y):
    return sqrt(x ** 2 + y ** 2 )
 
cos = scalar(x1, y1, x2, y2)/(module(x1,y1)*module(x2,y2))
 
ang = acos(cos)
 
print(f'угол между векторами = {degrees(acos(cos))}')
print(f'Расстояние между точками = {sqrt( (x2 - x1)**2 + (y2 - y1)**2)}')
```
