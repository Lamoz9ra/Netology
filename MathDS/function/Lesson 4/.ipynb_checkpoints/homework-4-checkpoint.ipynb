### Задание 1

Придумайте функции со следующими свойствами (в a) b) c) область определения X любая, какая вам удобна, главное - функция с нужными свойствами):

a) бесконечное количество локальных экстремумов

b) 2 локальных экстремума

с) 3 локальных экстремума

d*) область определения функции - 1 точка

a) $f(x) = sin(x)$

b) $f(x) = x^3-2*x^2-5x+6$ - вроде как производная будет на степень ниже, а значит корней равных нулю степени N-1, что и говорит о лок экстремумах

c) $f(x) = x^4-2*x^3-5x^2+6x+2$

d) $f(x) = \sqrt{-x^2}$

```python
def fc(x):
    return x**4-7*x**3-5*x**2-6*x

x0 = -5 # start from x = -5
xmin0 = fmin(fc,x0)

x1 = -4 # start from x = -4
xmin1 = fmin(fc,x1)

# plot the function
x = np.linspace(-15,15,100)
y = fc(x)
pylab.plot(x,y)
# plot of x0 and the minimum found startin from x0
pylab.plot(x0, fc(x0), 'bd', xmin0, fc(xmin0), 'bo')
# plot of x1 and the minimum found startin from x1
pylab.plot(x1, fc(x1), 'rd', xmin1, fc(xmin1), 'ro')

pylab.axis([-13,13,-1200,1000])
pylab.show()
```

```python
def fb(x):
    return x**3-5*x**2-5*x+6

x0 = -5 # start from x = -5
xmin0 = fmin(fb,x0)

x1 = -4 # start from x = -4
xmin1 = fmin(fb,x1)

# plot the function
x = np.linspace(-15,15,100)
y = fb(x)
pylab.plot(x,y)
# plot of x0 and the minimum found startin from x0
pylab.plot(x0, fb(x0), 'bd', xmin0, fb(xmin0), 'bo')
# plot of x1 and the minimum found startin from x1
pylab.plot(x1, fb(x1), 'rd', xmin1, fb(xmin1), 'ro')

pylab.axis([-10,10,-200,200])
pylab.show()
```

### Задание 3:

Используя numpy, посчитайте значение 3-ей и 4-ой производных функций в точке:

$f(x) = x^5+4sin(2x)+cos(3x+3)$ в точке $x=1$


```python
from sympy import *

x = Symbol('x')
y = x**5+4*sin(2*x)+cos(3*x+3)

num=3 #Номер производной

for i in range(1,num+1,1):
    yprime = y.diff(x)
    y=yprime
    print(i,yprime)


# f = lambdify(x, yprime, 'numpy')
# f(np.ones(5))
```

```python
from scipy.misc import derivative
def f(x):
    return 60*x**2+27*np.sin(3*x+3)-32*np.cos(2*x)
derivative(f, 1.0, dx=1e-6)
```

```python
x = np.linspace(-10, 10)
fx = f(x)
f1x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]
```

```python
plt.plot(x, fx, label='function')
plt.plot(x, f1x, label='derivative')
plt.legend()
plt.show()
```
